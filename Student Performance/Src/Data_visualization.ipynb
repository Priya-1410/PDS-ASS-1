```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```


#1. Distribution of Math Scores by Gender:
# The histogram shows the distribution of math scores for both males and females.
# This makes it easy to compare the spread and central tendencies of scores across genders, revealing whether one group consistently scores higher or has a broader range of scores.

```python
# Step 2: Data Visualization

# Setting visual style
sns.set(style="whitegrid")

# 2.1 Distribution of Math Scores by Gender
plt.figure(figsize=(8, 6))
sns.histplot(data, x="math score", hue="gender", element="step", kde=True, palette="pastel")
plt.title("Distribution of Math Scores by Gender")
plt.savefig("/content/drive/MyDrive/Student_Performance/Result/math_scores_by_gender.png")
plt.show()
```

# 2. Average Subject Scores by Lunch Type:
# The bar chart visualizes the average scores in math, reading, and writing for students based on whether they receive standard or free/reduced lunch.
# It highlights the potential influence of socioeconomic factors (as represented by lunch type) on academic performance across all subjects.

```python
# 2.2 Average Subject Scores by Lunch Type
plt.figure(figsize=(8, 6))
lunch_scores = data.groupby('lunch')[['math score', 'reading score', 'writing score']].mean().reset_index()
lunch_scores = pd.melt(lunch_scores, id_vars="lunch", var_name="Subject", value_name="Average Score")
sns.barplot(data=lunch_scores, x="lunch", y="Average Score", hue="Subject", palette="muted")
plt.title("Average Subject Scores by Lunch Type")
plt.savefig("/content/drive/MyDrive/Student_Performance/Result/average_scores_by_lunch.png")
plt.show()
```

# 3. Math Scores Distribution by Parental Education Level:
# The boxplot shows the distribution of math scores based on parental education levels.
# This makes it easier to observe any trends where higher parental education may correlate with higher student performance, and whether there are outliers or significant variation within each category.

```python
# 2.3 Boxplot of Math Scores by Parental Level of Education
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="parental level of education", y="math score", palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Math Scores Distribution by Parental Education Level")
plt.savefig("/content/drive/MyDrive/Student_Performance/Result/math_scores_by_parental_education.png")
plt.show()
```

# 4. Correlation Heatmap of Subject Scores:
# The heatmap visualizes the correlation between math, reading, and writing scores.
# This simplifies the task of identifying strong or weak correlations between different subjects, showing that scores in one subject tend to be positively associated with others.

```python
# 2.4 Heatmap of Correlations Between Subject Scores
plt.figure(figsize=(6, 5))
corr_matrix = data[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr_matrix, annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap of Subject Scores")
plt.savefig("/content/drive/MyDrive/Student_Performance/Result/correlation_heatmap.png")
plt.show()
```

# 5. Test Preparation Course Completion by Race/Ethnicity:
# The count plot displays the number of students who completed the test preparation course by race/ethnicity.
# It provides insight into the distribution of test preparation across different demographic groups, which may reveal disparities in access to or completion of such resources.

```python
# 2.5 Count Plot of Test Preparation Course Completion by Race/Ethnicity
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x="race/ethnicity", hue="test preparation course", palette="Set2")
plt.title("Test Preparation Course Completion by Race/Ethnicity")
plt.savefig("/content/drive/MyDrive/Student_Performance/Result/test_preparation_by_race.png")
plt.show()
```
